```python
import bs4
from bs4 import BeautifulSoup
```

```python
with open('211rank.html') as infile:
    html_doc = infile.read()
```

```python
html_doc[:1024]
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
myrows = soup.find_all("tr", {"class": "rankorder"})
```

```python
most_recent_nation = ""
most_recent_value = 0
for row in myrows:
    for td in row.children:
        if(type(td)==bs4.element.Tag):
            if td.has_attr('class') and td['class'][0] == 'region':
                for chld in td.children:
                    if(type(chld)==bs4.element.Tag):
                        most_recent_nation = chld.contents[0]
            if('$' in td.contents[0]):
                most_recent_value = td.contents[0]
                print(f"{most_recent_nation} has a PPP per capita of {most_recent_value}")
```

```python
for row in myrows:
    for td in row.children:
        if(type(td)==bs4.element.Tag):
            if('$' in td.contents[0]):
                print(td.contents[0])
            if td.has_attr('class') and td['class'][0] == 'region':
                print(td)
```
